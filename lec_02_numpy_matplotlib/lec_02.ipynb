```python
f = open('test.dat', 'wb')
f.write(bytes('строка', 'utf8'))
f.close()
```

```python
try:
    file = open("test.txt", "r")
    print(file.read())
    file.close()
except (FileNotFoundError, IOError):
    print('error occured')
```

```python
try:
    f = open("test1.dat", "rb")
    b = f.read(1)
    s = ""
    while True:
        b = f.read(1)
        if b == b'':
            break
        s += b.hex()
    print(s)
    f.close()
except IOError as e:
    print('error: {}'.format(e.args))
else:
    print('ok')
finally:
    print('finally')
```

```python
try:
    raise Exception('spam', 'eggs')
except Exception as inst:
    print(type(inst))    # the exception instance
    print(inst.args)     # arguments stored in .args
    print(inst)          # __str__ allows args to be printed directly,
                         # but may be overridden in exception subclasses
    x, y = inst.args     # unpack args
    print('x =', x)
    print('y =', y)
```

```python
with open('test.txt', 'r') as file:
    print(file.read())
```

```python
with open('test.dat', 'r', encoding="utf-8") as file:
    print(file.read())
```

# Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import numpy as np
import math

fig = plt.figure()

x_p = [i for i in range(10)]
y1_p = [i**2 for i in x_p]
y2_p = [math.sqrt(200 * i) for i in x_p]

graph1 = plt.plot(x_p, y1_p, x_p, y2_p)
plt.legend(['$x^2$', '$\sqrt{200x}$'])

print('Plot: ', len(graph1), graph1)

text1 = plt.text(0.5, 50.5, 'magic plots')
print('Text: ', type(text1))

grid1 = plt.grid(True) 

#plt.savefig('plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
